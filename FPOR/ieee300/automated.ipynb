```python
from ieee_300_sqp_bb import *

import numpy as np
import pandas as pd

import time

```

```python
funcao_objetivo = []
iterations = []
voltages = [] 
angles = [] 
shunts = []
trafos =[]
reactive_power = []
times = []

for number in range(1,26):

    path = f'CLPSO/{number}/'

    try:
        
        s = time.time()
        
        results = Sequential_Quadratic_Programming_Branch_and_Bound(path,2,2)
        
        e = time.time()
            
        funcao_objetivo.append(results['funcao_objetivo'])

        iterations.append(results['iterations'])
        
        voltages.append(results['voltages'])
        
        angles.append(results['angles'])
        
        shunts.append(results['shunts'])
        
        trafos.append(results['trafos'])
        
        reactive_power.append(results['reactive_power'])
        
        times.append(e-s)

    except:

        funcao_objetivo.append(-1)

        iterations.append(-1)
        
        voltages.append(-1)
        
        angles.append(-1)
        
        shunts.append(-1)
        
        trafos.append(-1)
        
        reactive_power.append(-1)
        
        times.append(-1)

    print(number, funcao_objetivo[-1])
```

```python
pd.DataFrame(funcao_objetivo).to_csv('funcao_objetivo.csv')
pd.DataFrame(iterations).to_csv('iterations.csv')
pd.DataFrame(voltages).to_csv('voltages.csv')
pd.DataFrame(angles).to_csv('angles.csv')
pd.DataFrame(shunts) .to_csv('shunts.csv')
pd.DataFrame(trafos) .to_csv('trafos.csv')
pd.DataFrame(reactive_power).to_csv('qpower.csv')
pd.DataFrame(times).to_csv('times.csv')
```
